```python
#make a list of words in an article
from urllib import request
from bs4 import BeautifulSoup
from collections import Counter 
import spacy
url = "https://www.theguardian.com/commentisfree/2019/jun/24/boris-johnson-prime-minister-tory-party-britain"
```

```python
html = request.urlopen(url).read().decode('utf8')
```

```python
soup = BeautifulSoup(html, 'html.parser')
# soup.get_text()
```

```python
articleBody = soup.find("div", attrs=dict(itemprop="articleBody"))
```

```python
text = articleBody.getText()
```

```python
print(text)
```

```python
len(text)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
tokens = [t.text for t in doc if t.is_alpha and not t.is_stop]
```

```python
counts = Counter(tokens)
```

```python
counts
```

```python
result = filter(lambda token: len(token) == 5, tokens)
                                                    
```

```python
result_list = list(result) 
```

```python
print(result_list)
```

```python
most_common = counts.most_common(20)
```
